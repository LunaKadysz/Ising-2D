## Cadena cerrada 

Parto del Hamiltoniano:
$$ H_{N} = -J \sum_{i=1}^{N} \sigma_{i} \sigma_{i+1} - \mu B \sum_{i=1}^{N} \sigma_i $$ Simetrizo H 
$$ \longrightarrow H_{N} = -J \sum_{i=1}^{N} \sigma_{i} \sigma_{i+1} - \frac{\mu B}{2} \sum_{i=1}^{N} ( \sigma_{i} + \sigma_{i+1}) $$
Si definimos la matriz de transferencia q (), con $\lambda$ los autovals. Entonces nos queda que: 
$$ Q_N = \sum_{ {\sigma_j} } e^{\beta H_N} = ... = \sum_{\sigma_1} <\sigma_1 | q^N |\sigma_1> = tr(q^N) = \lambda_+ + \lambda_- $$

```python
#matriz de transferencia
q = Matrix([[exp(beta*(J+mu*B)), exp(-beta*J)], [exp(-beta*J), exp(beta*(J-mu*B))]])
q
```

```python
# autovalores y los simplificados
simplify(q.eigenvals())
```

```python
#autvects
q.eigenvects()
```
